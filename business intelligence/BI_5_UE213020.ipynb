```python
import pandas as pd
import json
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv("GDP_2022.csv")
df
```

```python
json.loads(df.to_json())
```

```python
my_dict = dict()

for key in df:
    my_dict[key] = dict()
    for index,val in enumerate(df[key]):
        my_dict[key][index] = val

print(json.dumps(my_dict,indent=2))
```
