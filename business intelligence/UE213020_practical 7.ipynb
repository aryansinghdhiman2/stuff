## Practical 7: To generate SQL based reports using aggregate operations

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Plastic Waste Around the World.csv')
df
```

### 1.Total Plastic Waste by Main Source

```python
# Total plastic waste grouped by the main source
total_waste_by_source = df.groupby('Main_Sources')['Total_Plastic_Waste_MT'].sum().reset_index()
total_waste_by_source
```

### 2. Average Recycling Rate by Coastal Waste Risk Level

```python
# Average recycling rate grouped by coastal waste risk level
avg_recycling_rate_by_risk = df.groupby('Coastal_Waste_Risk')['Recycling_Rate'].mean().reset_index()
avg_recycling_rate_by_risk
```

### 3. Top 5 Countries by Per Capita Waste

```python
# Top 5 countries with the highest per capita plastic waste
top_5_per_capita_waste = df[['Country', 'Per_Capita_Waste_KG']].nlargest(5, 'Per_Capita_Waste_KG')
top_5_per_capita_waste
```

### 4. Average Waste and Recycling Rate by Country

```python
# Average total plastic waste and recycling rate grouped by country
avg_waste_recycling_by_country = df.groupby('Country')[['Total_Plastic_Waste_MT', 'Recycling_Rate']].mean().reset_index()
avg_waste_recycling_by_country
```

### 5. Sum of Plastic Waste by Coastal Waste Risk Level

```python
# Sum of plastic waste grouped by coastal waste risk level
sum_waste_by_risk_level = df.groupby('Coastal_Waste_Risk')['Total_Plastic_Waste_MT'].sum().reset_index()
sum_waste_by_risk_level
```
