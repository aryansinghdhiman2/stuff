## Importing Required Libraries

```python
import pandas as pd
```

### Dataset Overview:
**Title:** *Country_population_2022.csv*

**Source:** *World Bank Group*  
**Reference:** [World Bank - Population Data](https://data.worldbank.org/indicator/SP.POP.TOTL)

**Description:**  
This dataset presents the population figures for various countries around the globe for the year 2022. It is structured as a two-dimensional dataset, providing key demographic data on a country-by-country basis.

**Columns:**
- **Country:** The name of each country included in the dataset.
- **Population:** The total population of the respective country as recorded in 2022.

```python
population_df = pd.read_csv("Country_population_2022.csv")
population_df
```

### Dataset Overview:
**Title:** *GDP_2022.csv*

**Source:** *World Bank Group*  
**Reference:** [World Bank - GDP (current US$)](https://data.worldbank.org/indicator/NY.GDP.MKTP.CD)

**Description:**  
This dataset presents the GDP figures for various countries around the globe for the year 2022. It is structured as a two-dimensional dataset.

**Columns:**
- **Country:** The name of each country included in the dataset.
- **GDP:** The total GDP of the respective country as recorded in 2022.


```python
gdp_df = pd.read_csv("GDP_2022.csv")
gdp_df
```

## Joining the datasets

The datasets are joined on the key attribute *Country*

```python
merged_df = pd.merge(population_df,gdp_df,on="Country",how="inner")
merged_df
```
