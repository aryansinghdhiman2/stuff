```python
import pandas as pd
```

## Reading Dataset

```python
college = pd.read_csv("college.csv",dtype=str)
college
```

```python
(college.isnull().sum())/len(college) * 100
```

```python
college.dropna(how='all',inplace=True)
(college.isnull().sum())/len(college) * 100
```

```python
df = college.drop('Non-Need-Based Aid+',axis=1)
df.head()
```

### From all marks based attributes, remove % sign from different attributes. In SAT/ACT column, Keep higher value.

```python
def removePercent(item:pd.Series):
    item.iat[4]=item.iat[4].rstrip("%")
    item.iat[5]=item.iat[5].rstrip("%")
    item.iat[7]=item.iat[7].rstrip("%")
    item.iat[8]=item.iat[8].rstrip("%")
    item.iat[12]=item.iat[12].rstrip("%")
    item.iat[13]=item.iat[13].rstrip("%")

    marks:list[str] = item.iat[5].split("/")
    if(len(marks)==2):
        item.iat[5] = marks[0] if int(marks[0].rstrip("*")) > int(marks[1].rstrip("*")) else marks[1]

df.apply(removePercent,axis=1)
df
```

### From all cost based attributes, Remove $ sign and commas.

```python
10,11,14,15

def removeDollar(item:pd.Series):
    item.iat[10]=item.iat[10].lstrip("$").replace(",","")
    item.iat[11]=item.iat[11].lstrip("$").replace(",","")
    item.iat[14]=item.iat[14].lstrip("$").replace(",","")
    item.iat[15]=item.iat[15].lstrip("$").replace(",","")

df.apply(removeDollar,axis=1)
df
```
